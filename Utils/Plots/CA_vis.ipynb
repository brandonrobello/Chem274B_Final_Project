```python
import matplotlib.pyplot as plt
import numpy as np

# Read the txt file and extract the dimensions and frames
with open('/home/radhikasahai/chem_274B/Chem274B_Final_Project/Bin/Test_CA_output.txt', 'r') as file:
    lines = file.readlines()
    dim_line = lines[0].split('\t')
    dim1, dim2 = int(dim_line[0]), int(dim_line[1])

    frames = []
    for line in lines[1:]:
        frame = [int(cell) for cell in line.split('\t')[:-1]]  # Exclude the last empty element
        frames.append(frame)

# Reshape each frame into a 2D matrix
frames_matrix = np.array(frames).reshape(len(frames), dim1, dim2)

# Create a visualization for each frame
for i, frame in enumerate(frames_matrix):
    plt.imshow(frame, cmap='binary', interpolation='nearest')
    plt.title(f'Frame {i+1}')
    plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np

# Read the txt file and extract the dimensions and frames
with open('/home/radhikasahai/chem_274B/Chem274B_Final_Project/Bin/Test_CA_output.txt', 'r') as file:
    lines = file.readlines()
    dim_line = lines[0].split('\t')
    dim1, dim2 = int(dim_line[0]), int(dim_line[1])

    frames = []
    for line in lines[1:]:
        frame = [int(cell) for cell in line.split('\t')[:-1]]  # Exclude the last empty element
        frames.append(frame)

# Reshape each frame into a 2D matrix
frames_matrix = np.array(frames).reshape(len(frames), dim1, dim2)

# Set up the figure and axis
fig, ax = plt.subplots()
img = ax.imshow(frames_matrix[0], cmap='binary', interpolation='nearest')

# Update function for animation
def update(frame):
    img.set_array(frames_matrix[frame])
    ax.set_title(f'Frame {frame+1}')
    return img,

# Create animation
animation = FuncAnimation(fig, update, frames=len(frames_matrix), interval=200, repeat=True, repeat_delay=1000)

# Save the animation to a file (adjust the filename and format as needed)
animation.save(animation.save('general_CA.gif', writer='ffmpeg', fps=5))

# Show the plot (optional)
plt.show()
```

```python
#test for Parity rule
# Read the txt file and extract the dimensions and frames
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap


# Define the color map
colors = ['white', 'palegreen', 'green', 'red', 'darkgreen']
cmap = ListedColormap(colors)

# Read the data from the text file
with open('/home/radhikasahai/chem_274B/Chem274B_Final_Project/Bin/Test_CA_ParityRule_output.txt', 'r') as f:
    lines = f.readlines()

# Process the data and reverse the order of frames
frames = []

m = {0:0,1:1,4:2,11:3,22:4} #hash to normalize outputted txt data to values for colormap
for line in lines[1:]:  # Skip the header
    frame = list(map(int, line.split()))
    
    normalizedFrame = []
    for value in frame: 
        normalizedFrame.append(m[value])

    frames.append(normalizedFrame)  # Insert frames at the beginning of the list

# Create a grid visual for each frame
for i, frame in enumerate(frames):
    grid = np.array(frame).reshape(15,15)  # Reshape the frame into a 9x9 grid 
    plt.figure()
    plt.imshow(grid, cmap=cmap, aspect='auto',vmin=0, vmax=4)  # Use the color map
    plt.title(f'Frames in ParityRule {i + 1}')
    plt.show()
```
